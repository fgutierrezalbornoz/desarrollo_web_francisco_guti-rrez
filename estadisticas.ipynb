```python
import matplotlib.pyplot as plt

# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = 'Música', 'Deporte', 'Ciencias', 'Religión', 'Política', 'Tecnología', 'Juegos', 'Baile', 'Comida', 'Otro'
sizes = [np.random.randint(2, 15) for i in range(9)]
sizes.append(100 - np.sum(sizes))

fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

Gráfico 3


```python
import matplotlib.pyplot as plt
import numpy as np


mañana = [np.random.randint(1, 20) for i in range(12)]
mediodia = [np.random.randint(1, 30) for i in range(12)]
tarde = [np.random.randint(1, 40) for i in range(12)]

meses = [
    "Enero", "Febrero", "Marzo", "Abril", "Mayo", "Junio",
    "Julio", "Agosto", "Septiembre", "Octubre", "Noviembre", "Diciembre"
]
x = np.arange(len(meses))*3
width = 0.7  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width, mañana, width, label='Mañana')
rects2 = ax.bar(x, mediodia, width, label='Mediodía')
rects3 = ax.bar(x + width, tarde, width, label='Tarde')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_xticks(x, meses)
ax.set_xticklabels(meses, rotation=45)
ax.legend()

ax.bar_label(rects1, padding=3)
ax.bar_label(rects2, padding=3)
ax.bar_label(rects3, padding=3)
fig.tight_layout()

plt.show()
```
